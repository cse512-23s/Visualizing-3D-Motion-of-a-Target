```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('a3data.csv')
```

```python
brush = alt.selection_interval(
    encodings=['x'] # limit selection to x-axis (time) values
)

# dynamic query lines
Y_T = alt.Chart(df).mark_line().add_selection(
    brush
).encode(
    alt.X('T(s):T', title='Time'),
    alt.Y('Y(mm):Q', title='Signal of Y direction')
).properties(
    width=650,
    height=50
)

# scatter plot, modify opacity based on selection
Y_Z = alt.Chart(df).mark_line(point=True).encode(
    x='Y(mm):Q',
    y='Z(mm):Q',
    tooltip=['Y(mm):N', 'Z(mm):N', 'T(s):N'],
    opacity=alt.condition(brush, alt.value(0.75), alt.value(0.05))
).properties(
    width=650,
    height=400
)

alt.vconcat(Y_T, Y_Z).properties(spacing=5)
```
